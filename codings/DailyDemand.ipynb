### we want to get daily data demand across city so lets remove all other columns and sum the energy demand on daily basis

```python
daily = data.drop(['hour', 'dayofweek', 'month', 'year', 'dayofyear', 'Temperature', 'Humidity'], axis=1)
```

```python
daily
```

```python
# Now calculating daily energy demand 

daily_energy = daily.groupby('Timestamp', as_index=False)['Demand'].sum()
```

```python
daily_energy
```

```python
daily_energy.to_csv('daily_energy_demand.csv', index=False)
```
