```python
import numpy as np
import pandas as pd
```

```python
solar = pd.read_csv("solarEnergyGenerated2.csv")
```

```python
solar.head()
```

```python
solar = solar[['date', 'kWh']]
```

```python
solar.info()
```

```python
solar['date'] = pd.to_datetime(solar['date']).dt.date
```

```python
# Now calculating daily energy generated 

daily_energy = solar.groupby('date', as_index=False)['kWh'].sum()
```

```python
daily_energy
```

```python
# we want only 5 years date there fore accessing data of latest 5 years

# Filter data from 2018-03-15 onward
daily_energy = daily_energy[daily_energy['date'] >= '2018-03-16']
```

```python
daily_energy
```

```python
# Chnaging the date columns range

new_dates = pd.date_range(start='2020-01-01', end='2024-12-31', freq='D')
```

```python
daily_energy['date'] = new_dates
```

```python
daily_energy
```

```python
daily_energy.to_csv('daily_energy_generated.csv', index=False)
```
